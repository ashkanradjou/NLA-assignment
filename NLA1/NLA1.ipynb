```python
import numpy as np

def gram_schmidt_qr(A):
    m, n = A.shape
    Q = np.zeros((m, n))
    R = np.zeros((n, n))

    for j in range(n):
        v = A[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], A[:, j])
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]

    return Q, R


A = np.array([[2.0, 1.0, -1.0], [1.0, 1.0, 0.0], [ 0.0, 2.0, 1.0]])
Q, R = gram_schmidt_qr(A)

print("Q:")
print(Q)
print("R:")
print(R)
print("A:")
print(np.dot(Q,R))

```

```python
#B
def HilbertMatrix(n):
  H = [[0 for x in range(n)] for y in range(n)]
  for i in range(n):
    for j in range(n):
      H[i][j] = 1 / ( i + j + 1 )
  return H
n = 100
print(HilbertMatrix(n))
```

```python
#C

n = 12
A = HilbertMatrix(n)
Q, R = gram_schmidt_qr(np.array(A))

print("Q:")
print(Q)
print("R:")
print(R)
print("A:")
print(np.dot(Q,R))
```
